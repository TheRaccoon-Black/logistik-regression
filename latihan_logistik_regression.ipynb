# latihan logistik regression 

```python
import pandas as pd

#mengubah dataset menjadi dataframe
df = pd.read_csv("sample_data/Social_Network_Ads.csv")
```

```python
#melihat lima baris pertama
df.head()
```

```python
#melihat nilai kosong pada atribut
df.info
```

```python
#drop kolong yang tidak diperlukan
data = df.drop(columns=["User ID"])

#menjalankan proses one_hot_encoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
#memisahkan atribut dan label
predictions = ["EstimatedSalary", "Gender_Female", "Gender_Male", "Age"]
X = data[predictions] 
y = data["Purchased"]
```

```python
#lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns=X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

#membagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.02, random_state=1)
```

```python
from sklearn import linear_model

#latih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
#uji akurasi model
akurasi = model.score(X_test, y_test)
print("akurasi dari model adalah : ", akurasi)
```
